# Retreiving the Table from Website


```python
import pandas as pd
tables = pd.read_html("https://rate.bot.com.tw/xrt?Lang=en-US")
# table = tables[7]
table = tables[0]
table = table.iloc[:, :5]
table.columns = ["Currency", "Cash Buying", "Cash Selling", "Spot Buying", "Spot Selling"]
table.set_index('Currency')
```

# Data Cleaning

```python
table.shape[0]
```

```python
# find the last instance of country code
import re

def get_curr_code(x):
    pat = re.compile('[A-Z]{3}')
    curr_code = pat.findall(x)
    return curr_code[0]

table["Currency"] = [get_curr_code(x) for x in table["Currency"]]

print(table)
```
